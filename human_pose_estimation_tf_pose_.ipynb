```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from tf_pose.estimator import TfPoseEstimator
from tf_pose.networks import get_graph_path, model_wh
import os
```

```python
model='cmu'
resize='656x368'
w, h = model_wh(resize)
```

```python
!ls
!cd /content/tf-pose-estimation/models/graph/cmu
!bash /content/tf-pose-estimation/models/graph/cmu/download.sh
```

```python
estimator = TfPoseEstimator(get_graph_path(model), target_size=(w, h))
```

```python
def plot_img(image):
    plt.imshow(image)
    plt.show()
```

```python
def get_human_pose(image):
    humans = estimator.inference(image, resize_to_default=(w > 0 and h > 0), upsample_size=4.0)
    image = TfPoseEstimator.draw_humans(image, humans)
    return image, humans
```

```python
def read_image(image_path):
  input_image=cv.imread(image_path)
  input_image=cv.cvtColor(input_image,cv.COLOR_BGR2RGB)
  return input_image
```

```python
image_path = '/content/drive/MyDrive/human_pose_estimation/tf-pose-estimation/images/p1.jpg'
input_image=read_image(image_path)
img, hum = get_human_pose(input_image)
plot_img(img)
```

```python
dir="/content/drive/MyDrive/human_pose_estimation/"
lst=[dir+file for file in os.listdir(dir)]
new_name=0
for file in lst:  
  if "png" in file or "jpg" in file or "PNG" in file or "JPG" in file:
    input_image=read_image(file)
    output_image,human=get_human_pose(input_image)
    plot_img(output_image)
    out_image=cv.cvtColor(output_image,cv.COLOR_BGR2RGB)
    file_dir="/content/drive/MyDrive/human_pose_estimation/results/"+str(new_name)+".jpg"
    cv.imwrite(file_dir,out_image)
    new_name+=1
```
